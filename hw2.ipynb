(a).t-SNE (and other dimensional reduction methods) are classified as unsupervised learning because it does not use labelled data in the training. Instead, it relies solely on the relationships between data points to create a meaningful lower-dimensional representation, which discovers the structure naturally from the dataset.

(a). The data given to t-SNE do not have labels because it is unsupervised, while hw1 is supervised training.

(b). X is the input data, an N*D numpy array. D is the original dimension of data, N is the number of datasets used for training. 

no_dims is the traget number of dimension the tsne reduces to. 

The tsne function first run PCA to reduce the dimension of X: D to initial_dims. PCA helps remove noise by projecting data onto the most informative components, which makes it easier for t-SNE to focus on meaningful local structure.

The perplexity means how many neighbors are considered significant. perplexity=2^{H(P)}, where H(P) is the Shannon entropy of the probability distribution over neighbors.

The function sets no_dims=2, initial_dims=50, perplexity=30.0 by default.

(f). In t-SNE, for each pair of $x_i$ and $x_j$ in the high dimensional space, the similarity between them is
$p_{ij}=e^{-d(x_i,x_j)^2/2\sigma_i^2}/\sum_{ij}{e^{-d(x_i,x_j)^2/2\sigma_i^2}}$, where $d(x_i,x_j)$ is the Eucidean distance, and hre it is defined as $|x_i-x_j|$. This is because Ising model is local, the probability only depends on Eucidean distance, i.e. if two points are very far away, they have small effects on each other.

However, in classical gauge theory, the phase is classified by global topological differences. We may have two configurations in the same phase but are distant in Euclidean distance. This is why t-SNE fails to classify it.
